## 3. Linear Regression Model (for Miles Per Gallon prediction)
### Model Training Using Split Data

```python
X = data[['cylinders', 'displacement', 'horsepower', 'weight', 'acceleration']]
y = data[['mpg']]
```

`The split sizes are tested in descending order to see the trend in accuracy increase over each size`

```python
sizes = [0.25, 0.20, 0.15]
maxr2 = 0

for size in sizes:
    for i in range(0, 10000):    
        
        X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=size, random_state=i)
        model = LinearRegression()
        model.fit(X_train, Y_train)
        y_pred = model.predict(X_test)
        
        r2 = r2_score(Y_test, y_pred)
        
        if i % 1000 == 0:
            print(f'--------------------------------------------------------------------- Size {size} Iteration {i} ')
        
        if r2 > maxr2:
            print(f'R2 {round(r2, 8)} -> Train Size {size*100}%  Random State: {i}')
            maxr2 = r2
```

`Since a smaller test size gives higher goodness of fit, let's try minimum size with multiple iterations of random state to produce best possible fit`



```python
X = data[['cylinders', 'displacement', 'horsepower', 'weight', 'acceleration']]
y = data[['mpg']]
```

```python
size = 0.05
maxr2 = 0
for i in range(0, 100_000):    
    
    X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=size, random_state=i)
    model = LinearRegression()
    model.fit(X_train, Y_train)
    y_pred = model.predict(X_test)
    
    r2 = r2_score(Y_test, y_pred)
    
    if i % 5000 == 0:
        print(f'--------------------------------------------------------------------- Size {size} Iteration {i} ')
    
    if r2 > maxr2:
        print(f'R2 {round(r2, 8)} -> Train Size {size*100}%  Random State: {i}')
        maxr2 = r2
```

`Best Training Settings -> Size : 0.05, Random State : 64013`

## 4. Naive Bayes Classifier (for Country of Origin prediction)

`Categorical to quantitative conversion`

```python
def origin_map(x):
    if x == 'usa':
        return 0
    elif x == 'japan':
        return 1
    else:
        return 2
```

```python
data['country'] = [origin_map(value) for value in data['origin']]
```

```python
data.head()
```

```python
X = data[['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration']]
y = data['country']
```

```python
sizes = [0.25, 0.20, 0.15, 0.10]
max_acc = 0

for size in sizes:
    for i in range(0, 10_000):
        
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=size, random_state=i)
        nbmodel = GaussianNB()
        nbmodel.fit(X_train, y_train)
        
        y_pred = nbmodel.predict(X_test)
        acc = accuracy_score(y_test, y_pred)
        
        if acc > max_acc:
            print(f'Accuracy -> {round(acc, 5)} SplitSize {size} Random State {i}')
            max_acc = acc
            
        
            
        if i % 1000 == 0:
            print(f'------------------------------------------------------------------ Size {size} Iteration {i}')
```

`Again we see that a smaller spilt gives better accuracy and precision, let's try and optimize it`

```python
sizes = [0.1, 0.05]
max_acc = 0

while max_acc < 1:
    for size in sizes:
        for i in range(0, 100_000):

            X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=size, random_state=i)
            nbmodel = GaussianNB()
            nbmodel.fit(X_train, y_train)

            y_pred = nbmodel.predict(X_test)
            acc = accuracy_score(y_test, y_pred)

            if acc > max_acc:
                print(f'Accuracy -> {round(acc, 8)} SplitSize {size} Random State {i}')
                max_acc = acc    

            if i % 5000 == 0:
                print(f'---------------------------------------------------------------- Size {size} Iteration {i}')
```

`Best Train Settings -> Split 0.05  Random State 2390`
